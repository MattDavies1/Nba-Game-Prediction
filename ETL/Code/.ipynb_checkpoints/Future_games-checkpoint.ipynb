```python
import pandas as pd
```

```python
future_games = pd.read_excel("../Resources/future_games.xlsx")
future_games
```

```python
team = pd.read_csv("../Resources/teams.csv")
team = team[["TEAM_ID","CITY"]]
team
```

```python
new_df = future_games.drop(columns=[
       'Unnamed: 1', 'Unnamed: 3',])
new_df.loc[new_df["Home/Neutral"]=="New York Knicks"]
```

```python
clean_df= new_df.rename(columns={'Visitor/Neutral':"CITY"})
```

```python
merge = pd.merge(clean_df,team,on="CITY",how="inner")
merge
```

```python
new = merge.rename(columns={"CITY":"Visitor","Home/Neutral":"CITY"})
```

```python
merge2 = pd.merge(new,team,on="CITY",how="inner")
merge2
```

```python
clean_df =merge2[['Date','Visitor','TEAM_ID_y','CITY','TEAM_ID_x']]
```

```python
clean_df
```

```python
clean_df.sort_values("Date",ascending=False)
```

```python
clean_df['Date'] = clean_df['Date'].astype('datetime64[ns]')
```

```python
new_df= clean_df.sort_values("Date",ascending=True)
```

```python
new_df= new_df[["Date","CITY","TEAM_ID_y","Visitor","TEAM_ID_x"]].reset_index()
new_df
```

```python
new_df = new_df.drop(labels="index",axis=1)
```

```python
new_df = new_df.rename(columns={"TEAM_ID_y":"Home_id","TEAM_ID_x":"Away_id","CITY":"Home","Visitor":"Away"})
```
